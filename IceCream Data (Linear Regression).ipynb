```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\admin\Desktop\Study Material\SEM V\Machine Learning\IceCreamData.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:, 0:-1].values
Y = data.iloc[:, -1].values

X.shape, Y.shape
```

```python
plt.scatter(X, Y)
plt.title('Temperature vs Revenue')
plt.xlabel('Temerature')
plt.ylabel('Revenue')
plt.show()
```

```python
# Split the data into training and testing sets (80% train, 20% test)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=100)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the model
regression = LinearRegression()

# Fit the models on the training data
regression.fit(X_train, Y_train)

# Predict on the test data
Y_pred = regression.predict(X_test)
```

```python
plt.scatter(X_test, Y_test, color='red', label='Actual Data')
plt.plot(X_test, Y_pred, color='blue', label='Predicted Curve')
plt.title('Temperature vs Revenue')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate Mean Squared Error
MSE = mean_squared_error(Y_test, Y_pred)
print(f'MSE : {MSE}')

# Calculate R-squared score
r_squ = r2_score(Y_test, Y_pred)
print(f'R-squared Score : {r_squ}')
```
