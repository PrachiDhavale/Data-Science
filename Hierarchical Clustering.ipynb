```python
# Importing essential liabraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
```

```python
# Reading a dataset
df = pd.read_csv(r'C:\Users\admin\Desktop\iris.csv')
df.head()
```

```python
df = df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
df.head()
```

```python
# Making clusers
clusters = hierarchy.linkage(df, method="single")
print(clusters)
```

```python
# Plotting DEndrogram
plt.figure(figsize=(8,6))
dendrogram = hierarchy.dendrogram(clusters)
```

```python
# Defining Number of clusters
clustering_model = AgglomerativeClustering(n_clusters=3, linkage="single")
clustering_model.fit(df)
labels = clustering_model.labels_
print(labels)
```

```python
# Visualizing Clusters
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
sns.scatterplot(ax=axes[0], data=df, x='sepal.length', y='sepal.width').set_title('Without cliustering')
sns.scatterplot(ax=axes[1], data=df, x='sepal.length', y='sepal.width', hue=clustering_model.labels_).set_title('With clustering');
```
