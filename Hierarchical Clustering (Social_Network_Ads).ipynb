
# Hierarchical Clustering

```python
# Importing essential liabraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
```

```python
# Exporting dataset
data = pd.read_csv(r'C:\Users\admin\Desktop\Study Material\SEM V\Machine Learning\Social_Network_Ads .csv')
data.head()
```

```python
# Encoding Feature : 'Gender'
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for column in data :
    if data[column].dtype=='object' :
        data[column] = le.fit_transform(data[column])
```

```python
# Taking independent variable in x
x = data[['Age', 'EstimatedSalary']]
x.head()
```

```python
# Taking Target variable in y
y = data[['Purchased']]
y.head()
```

```python
# Making clusers
clusters = hierarchy.linkage(x, method="single")
print(clusters)
```

```python
# Plotting DEndrogram
plt.figure(figsize=(8,6))
dendrogram = hierarchy.dendrogram(clusters)
```

```python
# Defining Number of clusters
clustering_model = AgglomerativeClustering(n_clusters=2, linkage="complete")
clustering_model.fit(x)
labels = clustering_model.labels_
print(labels)
```

```python
# Visualizing Clusters
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
sns.scatterplot(ax=axes[0], data=data, x='Age', y='EstimatedSalary').set_title('Without cliustering')
sns.scatterplot(ax=axes[1], data=data, x='Age', y='EstimatedSalary', hue=clustering_model.labels_).set_title('With clustering');
```
