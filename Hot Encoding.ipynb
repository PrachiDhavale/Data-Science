```python
import pandas as pd

# Load the dataset

df = pd.read_csv(r'C:\Users\admin\Desktop\Study Material\SEM V\Machine Learning\Housing .csv')

# Display the first few rows
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select the 'furnishingstatus' column
furnishingstatus = df[['furnishingstatus']]

# Initialize OneHotEncoder with the updated argument
encoder = OneHotEncoder(sparse_output=False)

# Apply the encoder to the 'furnishingstatus' column
encoded_furnishingstatus = encoder.fit_transform(furnishingstatus)

# Create a DataFrame with the encoded values and the original column names
encoded_df = pd.DataFrame(encoded_furnishingstatus, columns=encoder.get_feature_names_out(['furnishingstatus']))

# Concatenate the encoded columns back to the original dataframe
df_encoded = pd.concat([df, encoded_df], axis=1).drop(columns=['furnishingstatus'])

# Display the first few rows of the new dataframe
df_encoded.head()
```
