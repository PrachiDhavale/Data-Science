```python
# Categorize 'age' column into three different categories and save it in a new variable

def categorize(age) :
    if age < 30 :
        return 'Young'
    elif age < 50 :
        return 'Middle-aged'
    else :
        return 'Senior'


data['age_category'] = data['age'].apply(categorize)
```
