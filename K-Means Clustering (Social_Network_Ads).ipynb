```python
# Importing essential liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn import datasets
import mpl_toolkits.mplot3d
```

```python
# Exporting dataset
data = pd.read_csv(r'C:\Users\admin\Desktop\Study Material\SEM V\Machine Learning\Social_Network_Ads .csv')
data.head()
```

```python
data.isnull().sum()
```

```python
# Encoding Feature : 'Gender'
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for column in data :
    if data[column].dtype=='object' :
        data[column] = le.fit_transform(data[column])
```

```python
# Taking independent variable in x
X = data[['Age', 'EstimatedSalary']]
X.head()
```

```python
# Taking Target variable in y
y = data[['Purchased']]
y.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Apply K-Means with the chosen number of clusters
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
data['KMeans_Cluster'] = kmeans.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(8, 5))
plt.scatter(X['Age'], X['EstimatedSalary'], c=data['KMeans_Cluster'], cmap='viridis', edgecolors='k', alpha=0.75)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.xlabel('Age (Standardized)')
plt.ylabel('Estimated Salary (Standardized)')
plt.title('K-Means Clustering on Social_Network_Ads')
plt.legend()
plt.show()
 
```
